```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price' : [55000,20000,30000,2000,8000,45000],
    'Units_sold':[25,60,35,80,50,15]
}
df = pd.DataFrame(data)
df
```

```python
# Data Binning
bins1 = [0,10000,30000,60000]
labels1 = ['Low','Medium','High']

df['Price_category'] = pd.cut(df['Price'],bins=bins1,labels=labels1)
df
```

```python
# Binning Units Sold
bins2 = [0,30,60,100]
labels2 = ['Low Sales','Medium Sales','High Sales']

df['Sales_category'] = pd.cut(df['Units_sold'],bins=bins2,labels=labels2)
df
```

```python
# Data formating:
# Convert Data type

print(df.dtypes)
df['Units_sold'] = df['Units_sold'].astype(float)
print(df.dtypes)
```

```python
# Convert Text to upercase
df['Product'] = df['Product'].str.upper()
df
```

```python
# Rounding Values
df['Price'] = df['Price'].round(2)
df
```

```python
# Additional
# Sort Data
sorted_df = df.sort_values(by ='Price')
print(sorted_df)
```

```python
# Display Unique Values
print(df['Price_category'].unique())
```

```python
# Dataset: Food Delivery orders
import pandas as pd
import numpy as np

data = {
    'Order_ID':['01','02','03','04','05','06','07','08'],
    'Order_Value' : [250,450, 120, 600, 300, 150, 500, 220],
    'Delivery_Time' : [30,45,25, 60, 35, 20, 50, 28],
     'Distance_km' : [5,8,3,10,6,2,9,4]
}
df = pd.DataFrame(data)
df

```

```python
# Binning Order Values

bins = [0,200,400,700]
labels = ['Low order','Medium order','High order']

df['Order_category'] = pd.cut(df['Order_Value'],bins=bins,labels=labels)
print(df)
```

```python
# Binning Delivery time
bins2 = [0,30,45,60]
labels2 = ['Fast Delivery','Average Delivery','Slow Delivery']
df['Delivery_Time'] = pd.cut(df['Delivery_Time'], bins=bins2, labels=labels2)
print(df)
```

```python
# Binning kilometer
bins3 = [0,5,8,10]
labels3 = ['Near','Medium','Far']
df['Distance_category'] = pd.cut(df['Distance_km'], bins=bins3, labels=labels3)
print(df)
```

```python
# Data formating:
# Convert Data type

print(df.dtypes)
df['Distance_km'] = df['Distance_km'].astype(float)
print(df.dtypes)
```

```python
# Count of categories
print(df['Order_category'].value_counts())
df
```

```python
# Sort Data
sorted_df = df.sort_values(by ='Order_Value')
print(sorted_df)
```

```python
# Convert Text to upercase
df['Distance_category'] = df['Distance_category'].str.upper()
df
```
